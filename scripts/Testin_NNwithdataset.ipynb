# TESTING NN WITH DATASETS

```python
from si.neural_netwoks.layer import *
from si.neural_netwoks.nn import NN
from si.data.dataset1 import Dataset
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from si.read_csv import read_csv
from si.model_selection.split import train_test_split
from si.metrics.cross_entropy import cross_entropy_derivate,cross_entropy
```

```python
dataset2= read_csv("C:/Users/Andr√© Silva/SI/datasets/breast-bin.csv", sep=",")
dataset2.X= StandardScaler().fit_transform(dataset2.X)
train,test= train_test_split(dataset2, test_size=0.30)
train.y
```

```python
n=train.X.shape[1]
n
```

```python
#Create a entry level layer

l1= Dense(input_size=n,output_size=n)
l2= Dense(input_size=n,output_size=n)
l3= Dense(input_size=n,output_size=1)

l1_sg= ReLUActivation()
l2_sg= ReLUActivation()
l3_sg= SigmoidActivation()
```

```python
nn_model= NN(layers=[l1,l1_sg,l2,l2_sg,l3,l3_sg],loss_derivate=cross_entropy_derivate,loss_function=cross_entropy)
```

```python
nn_model.fit(dataset=train)
```

```python
predict=nn_model.predict(dataset=train).reshape(1,-1)
```

```python
print(predict)
```

```python
predict= np.where(predict>0.5, 1,0).reshape(1,-1)
print(predict)
```
