# TESTING AND PLOTING Ridge and logistic

```python
from si.linear_model.logistic_regression import LogisticRegression
from si.linear_model.ridge_regression import RidgeRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from si.read_csv import read_csv
from sklearn import preprocessing
```

```python
#Ridge!

dataset= read_csv("C:/Users/Andr√© Silva/SI/cpu.csv", sep=",")
print(dataset.to_dataframe())
scalar= preprocessing.StandardScaler()
scalar.fit(dataset.X)
dataset.X=scalar.transform(dataset.X)

model= RidgeRegression(max_iter=2000)
model.fit(dataset)
print(model.score(dataset))
print(model.cost(dataset))
```

```python
import matplotlib.pyplot as plt 
ycol=[]
for keys in model.cost_history.keys():
    ycol.append(keys+1)
ycol= np.array(ycol)
xcol=[item for item in model.cost_history.values()]
xcol=np.array(xcol)
plt.plot(ycol,xcol)
```

```python
for value in model.cost_history.keys():
        print(f"cost of {value+1} iteration is: {model.cost_history[value]}")
```
